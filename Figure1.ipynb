```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import sklearn.model_selection as sks
import sklearn.metrics as skm
```

```python
x = np.arange(0, 6*np.pi, 0.01)
```

```python
p = [1, 0]
```

```python
sigma = 1
```

```python
noise1 = np.random.randn(x.shape[-1]) * sigma
```

```python
y1 = np.polyval(p, x) + noise1
```

```python
noise2 = np.random.randn(x.shape[-1]) * sigma
```

```python
y2 = np.polyval(p, x) + noise2
```

```python
plt.plot(y1, y2, '.')
```

```python
plt.plot(x, y1, '.')
```

```python
p_hat_1 = np.polyfit(x, y1, 1)
p_hat_2 = np.polyfit(x, y2, 1)
```

```python
y_hat_1 = np.polyval(p_hat_1, x)
y_hat_2 = np.polyval(p_hat_2, x)
```

```python
plt.plot(y1, y_hat_1, '.')
```

```python
plt.plot(y2, y_hat_2, '.')
```

```python
residuals1 = y1 - y_hat_1
residuals2 = y2 - y_hat_2
```

```python
plt.plot(residuals1, residuals2, '.')
```

```python
fig, ax = plt.subplots(1)
ax.plot(noise1, residuals1, '.')
fig.savefig('./figures/figure1.png')
```

```python
skm.r2_score(y1, y_hat_1), skm.r2_score(y2, y_hat_2)
```
